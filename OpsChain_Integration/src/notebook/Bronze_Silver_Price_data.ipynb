```python
from pyspark.sql.types import *
from pyspark.sql.functions import * 
```

```python
df= spark.read.table('fmcg2.bronze.gross_price')
```

```python
display(df.limit(5))
```

```python
df.printSchema()
```

Normalize month field



```python
df.select('month').distinct().show()
```

```python
df=df.withColumn('month',coalesce(
    try_to_date(col('month'),'yyyy/mm/dd'),
    try_to_date(col('month'),'dd/mm/yyyy'),
    try_to_date(col('month'),'yyyy-mm-dd'),
    try_to_date(col("month"), 'dd-MM-yyyy')
))
```

```python


df = df.withColumn(
    "gross_price",
    when(col("gross_price") == "unknown", 0)
    .when(col("gross_price") == "not_available", 0)
    .otherwise(col("gross_price"))
)

display(df)
```

```python
df=df.withColumn('gross_price',regexp_replace(col("gross_price"),'-',''))
```

```python

df.display()
```

```python
df=df.withColumnsRenamed({'product_id':'product_code',
                          'gross_price':'price_inr',
                          'month':'year'})
```

```python
display(df.limit(5))
```

```python
df.printSchema()
```

```python
df.printSchema()
```

```python
df.write.format('delta')\
    .mode('overwrite')\
    .saveAsTable('fmcg2.silver.price_data_enr')
```

```python
df.write.format('delta')\
    .option('overwriteSchema','true')\
    .mode('overwrite')\
    .saveAsTable('fmcg2.gold.price_data_gold')
```
